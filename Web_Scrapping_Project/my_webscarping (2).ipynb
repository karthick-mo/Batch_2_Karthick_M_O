```python
import requests

from bs4 import BeautifulSoup

r = requests.get("https://www.amazon.in/s?k=poco+f7&crid=3CZKO1LHO52LU&sprefix=poco%2Caps%2C587&ref=nb_sb_ss_mvt-t11-ranker_1_4")
c = r.content
c
```

```python
soup=BeautifulSoup(c,"html.parser")
```

```python
soup
```

```python
all=soup.find_all("div",{"class":"sg-col-inner"})
all
```

```python
for item in all:   #to display all price value in the website
    print(item.find_all("div",{"class":"KzDlHZ"})[0].text)
    print(item.find_all("div",{"class":"Nx9bqj _4b5DiR"})[0].text)
    print(item.find_all("ul",{"class":"G4BRas"})[0].text)
```

```python
l=[]
for item in all:   #to display all price value in the website
    d={}
    d["phone_name"]=item.find_all("div",{"class":"KzDlHZ"})[0].text
   
    d["phone_price"]=item.find_all("div",{"class":"Nx9bqj _4b5DiR"})[0].text.replace("₹","Rs.")
    d["phone_description"]=item.find_all("ul",{"class":"G4BRas"})[0].text
    l.append(d)
```

```python
l
```

```python
import pandas
df=pandas.DataFrame(l)
```
